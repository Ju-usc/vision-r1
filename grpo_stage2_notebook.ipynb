```python
# Import necessary libraries
import torch
from PIL import Image
from transformers import AutoTokenizer, AutoProcessor, Qwen2_5_VLForConditionalGeneration
from google.colab import files
import matplotlib.pyplot as plt

# Load model from Hugging Face
model_id = "juyounglee/qwen-vl-recipe-generator_v2_ex80"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-3B-Instruct")

# System prompt for GRPO
SYSTEM_PROMPT = """Look at the food image and create a recipe in following XML format:

<think>think step by step to infer the recipe from the image</think>
<recipe>
  <title>Name of the dish</title>
  <ingredients>
    <ingredient>Quantity and ingredient (e.g., 2 tbsp olive oil)</ingredient>
    <!-- Add more ingredients as needed -->
  </ingredients>
  <instructions>
    <step>1. Preheat the oven to 350°F.</step>
    <!-- Add more steps as needed -->
  </instructions>
</recipe>

Format requirements:
- Follow the exact XML structure shown above
- List each ingredient in a separate <ingredient> tag with quantity
- Present each cooking step in a separate <step> tag
- Use clear, concise language throughout
- Do not generate any text outside of the <recipe> tags and <think> tags
"""
```

```python
import os
jpg_lookup = {
    os.path.splitext(fname)[0]: os.path.join(food_images_dataset_path, fname)
    for fname in os.listdir(food_images_dataset_path)          # O(#files)
    if fname.endswith(".jpg")                  # fast string check
}
```

```python
def attach_jpg(ex):
    return {"image": jpg_lookup.get(ex["Image_Name"])}  # gives None if missing
```

```python
hf_dataset = hf_dataset.map(attach_jpg, num_proc=os.cpu_count())
hf_dataset = hf_dataset.cast_column("image", Image())  # lazy‐loads PIL.Image
```

```python
from PIL import Image
def get_image_path(metadata_entry):
    image_path = metadata_entry['image_path']
    return os.path.join(food_images_dataset_path, image_path.replace("\\", "/"))
```

```python
def generate_response(image):
  messages = [
  {"role": "system", "content": [{"type": "text", "text": SYSTEM_PROMPT}]},
  {"role": "user", "content": [
      {"type": "image", "image": image},
      {"type": "text", "text": "Generate a recipe based on this food image."}
  ]}]
  text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
  images = [content["image"] for message in messages for content in message.get("content", [])
            if content.get("type") == "image"]
  inputs = processor(text=[text], images=images, padding=True, return_tensors="pt").to(model.device)

  # Generate and decode
  generated_ids = model.generate(**inputs, max_new_tokens=1024, temperature=0.7, do_sample=True)
  generated_ids_trimmed = generated_ids[0][len(inputs.input_ids[0]):]
  recipe = tokenizer.decode(generated_ids_trimmed, skip_special_tokens=True)
  return recipe
```

```python
from transformers import pipeline

BASELINE_SYSTEM_PROMPT = """Look at the food image and create a recipe in following XML format:
<recipe>
  <title>Name of the dish</title>
  <ingredients>
    <ingredient>Quantity and ingredient </ingredient>
    <ingredient>Another ingredient with quantity</ingredient>
    <ingredient>A third ingredient with quantity</ingredient>
  </ingredients>
  <instructions>
    <step>1. First instruction step with clear details</step>
    <step>2. Second instruction step</step>
    <step>3. Third instruction step</step>
  </instructions>
</recipe>


Format requirements:
- Follow the exact XML structure shown above
- List each ingredient in a separate <ingredient> tag with quantity
- Present each cooking step in a separate <step> tag
- Use clear, concise language throughout
- Make sure to include both opening and closing tags for each element
- Do not generate any text outside of the <recipe> tags
"""
```

```python
  pipe = pipeline("image-text-to-text", model="Qwen/Qwen2.5-VL-3B-Instruct")

```

```python
def generate_response_baseline(img):
  messages = [
      {
          "role": "system",
          "content": [
              {"type": "text", "text": BASELINE_SYSTEM_PROMPT},
          ],
      },
      {
          "role": "user",
          "content": [
              {
                  "type": "image",
                  "url": img
              },
              {"type": "text", "text": "Look at the food image and create a recipe in XML format"},
          ],
      },
  ]
  response = pipe(
    text=messages,
    max_new_tokens=1024,
    temperature=1.0,
    return_full_text=False
  )
  return response[0]["generated_text"]
```

```python
eval_data_baseline = []

for sample in eval_dataset_metadata:
  image_path = get_image_path(sample)
  if os.path.exists(image_path):
      try:
          img = Image.open(image_path).convert("RGB")
          print(f"Generating for: {image_path}")
          response = generate_response_baseline(img)
          eval_data_baseline.append({
              "image_path": image_path,
              "response": response
          })

      except Exception as e:
          print(f"Error processing {image_path}: {e}")
  else:
      print(f"Image missing: {image_path}")
```

```python
from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=os.environ["OPENROUTER_API_KEY"],
)
```

```python
import base64, io
from PIL import Image

def generate_response_baseline_32B(img):
  img_data_url = encode_as_data_url(img)

  completion = client.chat.completions.create(
  extra_body={},
  model="qwen/qwen2.5-vl-32b-instruct:free",
  messages = [
      {
          "role": "system",
          "content": [
              {"type": "text", "text": BASELINE_SYSTEM_PROMPT},
          ],
      },
      {
          "role": "user",
          "content": [
              {
                  "type": "image_url",
                  "image_url": img_data_url
              },
              {"type": "text", "text": "Look at the food image and create a recipe in XML format"},
          ],
      },
  ])
  output = completion.choices[0].message.content
  # print(output)
  return output
def encode_as_data_url(pil_img: Image.Image) -> str:
    buf = io.BytesIO()
    pil_img.save(buf, format="JPEG")
    b64 = base64.b64encode(buf.getvalue()).decode()
    return f"data:image/jpeg;base64,{b64}"
```
